````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

````python
covid_19 = covid_df['COVID19']
plt.hist( covid_19, bins =7)
plt.xlabel('COVID-19 Deaths')
plt.ylabel('frequency')
plt.title('Distribution of COVID-19 Deaths with Underlying Health Conditions')
plt.show()
````

Here we can visualize that majority of COVID 19 deaths among individuals with underlying health conditions are at the lower end of the spectrum. Fewer underlying health conditions are less likely to experience lower counts of COVID-19 death.As the number of deaths increases, the frequency decreases significantly, with very few occurrences beyond 6,000 deaths.

````python
sns.pairplot(covid_df, hue='Dataasof');
plt.title('Pairplot of COVID-19 Deaths with Underlying Health Conditions')
plt.show()
````

The pairplot provides a visual summary of the relationships between the numerical variables of COVID-19 data. By carefully observing the patterns in the plots, we can see the insights, how different factors  related to COVID-19 deaths.The correlation between COVID-19 deaths and certain health condition becomes stronger. Also certain corelation become weaker over time.

The data shows that COVID-19 deaths are not uniformly distributed across age groups. Certain age groups, especially the young(18-29 years), adults(60-64 years), and older adults(70-79 years)appear to be more severely affected.

````python

age_group_deaths = covid_df.groupby('AgeGroup')['COVID19'].sum()
plt.figure(figsize=(12, 6))
age_group_deaths.plot(kind='bar', color='skyblue')
plt.title('COVID-19 Death Counts by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

 This chart highlights the critical impact of age on COVID-19 mortality, showing that older individuals are at a higher risk of death from the virus.

````python
# Group data by month
monthly_deaths = covid_df.groupby("Date_Of_Death_Month")["Diabetes.mc"].sum()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_deaths.index, y=monthly_deaths.values, marker="o")
plt.title("Diabetes-Related Deaths Over Time (2020)")
plt.xlabel("Month")
plt.ylabel("Death Count")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
````

Diabetes-related deaths peaked in April 2020, aligning with COVID-19 surges.

````python

covid_df.plot(kind='scatter', x='Diabetes.mc', y='C19PlusDiabetes', s=32, alpha=.8)
plt.title('Relationship between Diabetes.mc and C19PlusDiabetes')
plt.gca().spines[['top', 'right',]].set_visible(False)
````

Here, the scatterplot appears to have a positive correlation(linear trend) between "Diabetes.mc"(diabetes measurements) and "C19PlusDiabetes"(COVID-19-related diabetes).The positive correlation suggests that individuals with higher diabetes measurements are more likely to have higher COVID-19-related diabetes cases.

````python

covid_df.groupby('Sex').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.title('Distribution of COVID-19 Deaths by Sex')
plt.xlabel('Count')
plt.ylabel('Sex')
plt.gca().spines[['top', 'right',]].set_visible(False)
````

 The bar for males is slightly longer than for females, indicating a higher count of COVID-19 deaths among males. Although males have a higher number of deaths, the difference is not extremely large. Both genders have a significant impact.

***For each age group, what is the percentage of COVID-19 deaths attributed to diabetes, hypertension, or other conditions?

````python
condition_columns = [
    'Diabetes.uc', 'Diabetes.mc', 'C19PlusDiabetes', 'C19PlusHypertensiveDiseases',
    'C19PlusMajorCardiovascularDiseases', 'C19PlusHypertensiveDiseasesAndMCVD',
    'C19PlusChronicLowerRespiratoryDisease', 'C19PlusKidneyDisease',
    'C19PlusChronicLiverDiseaseAndCirrhosis', 'C19PlusObesity'
]

# Calculate percentage
age_group_percentages = covid_df.groupby('AgeGroup')[condition_columns].sum()
age_group_percentages = age_group_percentages.div(age_group_percentages.sum(axis=1), axis=0) * 100

# Transpose
age_group_percentages = age_group_percentages.T

#  chart
fig, ax = plt.subplots(figsize=(12, 6))
age_group_percentages.plot(kind='bar', stacked=True, ax=ax)

plt.title('Percentage of COVID-19 Deaths Attributed to Conditions by Age Group')
plt.xlabel('Condition')
plt.ylabel('Percentage of Deaths')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Format percent sign
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0f}%'.format(y)))

plt.show()
````

***What is the relation between healthconditions: Diabetes, hypertension, and Cardiovascular disease?

````python

health_conditions = ["C19PlusDiabetes", "Diabetes.mc", "Diabetes.uc", "C19PlusHypertensiveDiseases", "C19PlusMajorCardiovascularDiseases", "C19PlusObesity"]  # Fixed: Separated column names and removed extra spaces/tabs
correlation_matrix = covid_df[health_conditions].corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Diabetes, Hypertension, and Other Conditions")
plt.show()
````

***What is the total number of deaths associated with diabetes in 2020?

````python
total_diabetes_deaths = covid_df['C19PlusDiabetes'].sum()
print(f"Total deaths associated with diabetes in 2020: {total_diabetes_deaths}")
````

````python


covid_df['Dataasof'] = pd.to_datetime(covid_df['Dataasof'])
covid_df['Month'] = covid_df['Dataasof'].dt.month

monthly_diabetes_deaths = covid_df.groupby('Month')['C19PlusDiabetes'].sum()

plt.figure(figsize=(10, 6))
monthly_diabetes_deaths.plot(kind='bar', color='skyblue')
plt.title('Monthly Variation in Diabetes-Related COVID-19 Deaths')
plt.xlabel('Month')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

***Are there gender-based differences in diabetes-related deaths?

````python

sex_diabetes_deaths = covid_df.groupby('Sex')['C19PlusDiabetes'].sum()
plt.figure(figsize=(8, 6))
sex_diabetes_deaths.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Diabetes-Related COVID-19 Deaths by Sex')
plt.xlabel('Sex')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

total_deaths = sex_diabetes_deaths.sum()
for sex, deaths in sex_diabetes_deaths.items():
    percentage = (deaths / total_deaths) * 100
    print(f"{sex}: {deaths} deaths ({percentage:.1f}%)")
````

***Is there any co relation between Diabetes and Hypertension in Covid19 deaths?

````python

health_conditions = ['C19PlusDiabetes', 'C19PlusHypertensiveDiseases']
subset_df = covid_df[health_conditions]


correlation_matrix = subset_df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Diabetes and Hypertension in COVID-19 Deaths")
plt.show()
correlation_coefficient = correlation_matrix.loc['C19PlusDiabetes', 'C19PlusHypertensiveDiseases']
print(f"Correlation coefficient between diabetes and hypertension: {correlation_coefficient:.2f}")
````

***Which variable has the strongest correlation with diabetes-related deaths?

````python
health_conditions = [
    'Diabetes.uc', 'Diabetes.mc', 'C19PlusDiabetes', 'C19PlusHypertensiveDiseases',
    'C19PlusMajorCardiovascularDiseases', 'C19PlusHypertensiveDiseasesAndMCVD',
    'C19PlusChronicLowerRespiratoryDisease', 'C19PlusKidneyDisease',
    'C19PlusChronicLiverDiseaseAndCirrhosis', 'C19PlusObesity'
]
relevant_columns = health_conditions

correlation_matrix = covid_df[relevant_columns].corr()


correlation_with_diabetes = correlation_matrix['C19PlusDiabetes'].drop('C19PlusDiabetes')
strongest_correlated_variable = correlation_with_diabetes.abs().idxmax()
print(f"The variable with the strongest correlation with diabetes-related deaths is: {strongest_correlated_variable}")
print(f"Correlation coefficient: {correlation_with_diabetes[strongest_correlated_variable]:.2f}")
````

***What is the average number of COVID-19 deaths per month for each underlying health condition?

````python
covid_df['Month'] = pd.to_datetime(covid_df['Dataasof']).dt.month

condition_column = [
    'Diabetes.uc', 'Diabetes.mc', 'C19PlusDiabetes', 'C19PlusHypertensiveDiseases',
    'C19PlusMajorCardiovascularDiseases', 'C19PlusHypertensiveDiseasesAndMCVD',
    'C19PlusChronicLowerRespiratoryDisease', 'C19PlusKidneyDisease',
    'C19PlusChronicLiverDiseaseAndCirrhosis', 'C19PlusObesity'
]

monthly_avg_deaths_by_condition = {}

for condition in condition_column:
    monthly_avg_deaths = covid_df.groupby(['Month'])[condition].mean().reset_index()
    monthly_avg_deaths_by_condition[condition] = monthly_avg_deaths
for condition_column, monthly_avg_deaths in monthly_avg_deaths_by_condition.items():
    print(f"Average COVID-19 Deaths per Month for {condition_column}:\n")
    print(monthly_avg_deaths)
    print("\n")
````
